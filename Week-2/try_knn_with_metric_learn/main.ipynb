```python
import numpy as np

from sklearn import preprocessing
from sklearn.datasets import load_boston
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score, KFold
```

```python
boston_data = load_boston()
```

```python
x = preprocessing.scale(boston_data.data)
y = boston_data.target

print(x[:10])
print(y[:10])
```

```python
p_linspace = np.linspace(start=1.0, stop=10.0, num=200)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
p_max, score_max = None, float("-inf")
for p in p_linspace:
    knn_regressor = KNeighborsRegressor(n_neighbors=5, weights='distance', metric='minkowski', p=p)
    scores = cross_val_score(knn_regressor, x, y, scoring='neg_mean_squared_error', cv=kf)
    scores_mean = np.mean(scores)
    print(f"p coeff: {p}\tscore: {scores_mean}")
    
    if scores_mean > score_max:
        score_max = scores_mean
        p_max = p
        
print(f"MAX SCORE: {score_max}, p coeff: {round(p_max, 1)}")
```
